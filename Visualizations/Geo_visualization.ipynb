```python
import pandas as pd
import numpy as np
from pathlib import Path
import plotly.express as pximport 
import hvplot.pandas
import holoviews as hv
from holoviews import opts
import matplotlib.pyplot as plt
import datetime as dt
```

```python
geo_visualization = pd.read_pickle("../prediction_oil_prices.pkl")
```

```python
df = geo_visualization.reset_index()
```

```python
curve_plot = df.plot(x='date', y='Prices',figsize=(20,10))
```

## USA Visualization

```python
# Adding Vlines (USA) to the plot 

usa_curve_plot = curve_plot

usa_curve_plot = df.plot(x='date', y='Prices',figsize=(20,10),)
usa_curve_plot.set_xlabel("Article Date")
usa_curve_plot.set_ylabel("Oil Prices")
usa_curve_plot.set_title("Article's with USA as the most frequest word")
usa_curve_plot.axvline(dt.datetime(2012, 6, 9), color='red')
usa_curve_plot.axvline(dt.datetime(2012, 11, 12), color='red')
usa_curve_plot.axvline(dt.datetime(2012, 12, 13), color='red')
usa_curve_plot.axvline(dt.datetime(2014, 3, 7), color='red')
usa_curve_plot.axvline(dt.datetime(2014, 4, 21), color='red')
usa_curve_plot.axvline(dt.datetime(2014, 9, 13), color='red')
usa_curve_plot.axvline(dt.datetime(2015, 4, 21), color='red')
usa_curve_plot.axvline(dt.datetime(2017, 10, 16), color='red')
usa_curve_plot.axvline(dt.datetime(2018, 6, 6), color='red')
usa_curve_plot.axvline(dt.datetime(2018, 6, 13), color='red')
usa_curve_plot.axvline(dt.datetime(2019, 1, 25), color='red')
plt.show()
```
