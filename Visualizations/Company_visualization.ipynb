```python
import pandas as pd
import numpy as np
from pathlib import Path
import plotly.express as pximport 
import hvplot.pandas
import holoviews as hv
from holoviews import opts
import matplotlib.pyplot as plt
import datetime as dt
```

```python
comp_visualization = pd.read_pickle("../prediction_oil_prices.pkl")
```

```python
df = comp_visualization.reset_index()
df
```

```python
curve_plot = df.plot(x='date', y='Prices',figsize=(20,10),)
```

## SHELL Visualization

```python
# Adding Vlines (SHELL) to the plot 

shell_curve_plot = curve_plot

shell_curve_plot.set_xlabel("Article Date")
shell_curve_plot.set_ylabel("Oil Prices")
shell_curve_plot.set_title("Article's with Shell as the most frequest word")
shell_curve_plot = df.plot(x='date', y='Prices',figsize=(20,10),)
shell_curve_plot.axvline(dt.datetime(2011, 8, 16), color='red')
shell_curve_plot.axvline(dt.datetime(2011, 8, 18), color='red')
shell_curve_plot.axvline(dt.datetime(2012, 5, 23), color='red')
shell_curve_plot.axvline(dt.datetime(2015, 2, 18), color='red')
plt.show()
```

## EXXON Visualization

```python
# Adding Vlines (EXXON) to the plot 

exxon_curve_plot = curve_plot

exxon_curve_plot.set_xlabel("Article Date")
exxon_curve_plot.set_ylabel("Oil Prices")
exxon_curve_plot.set_title("Article's with EXXON as the most frequest word")
exxon_curve_plot = df.plot(x='date', y='Prices',figsize=(20,10),)
exxon_curve_plot.axvline(dt.datetime(2011, 7, 28), color='m')
exxon_curve_plot.axvline(dt.datetime(2011, 8, 30), color='m')
exxon_curve_plot.axvline(dt.datetime(2016, 12, 14), color='m')
plt.show()
```
