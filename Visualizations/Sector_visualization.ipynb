# Sector Visualizations

```python
import pandas as pd
import numpy as np
from pathlib import Path
import plotly.express as pximport 
import hvplot.pandas
import holoviews as hv
from holoviews import opts
import matplotlib.pyplot as plt
import datetime as dt
```

```python
#read in the pickle file as a dataframe
df_visualization = pd.read_pickle("../prediction_oil_prices.pkl")
```

## Gas Visualization

```python
#reset index so there are two columns for date and prices
gas_df = df_visualization.reset_index()
```

```python
#read the column types
gas_df.dtypes
```

```python
#list the column values
list(gas_df.columns.values)
```

```python
#Create plot with vlines for the dates in which the word was most frequently used
gas_curve_plot = gas_df.plot(x='date', y='Prices', figsize=(20,10), legend=True)
gas_curve_plot.set_xlabel("Article Date")
gas_curve_plot.set_ylabel("Oil Prices")
gas_curve_plot.set_title("Article's with Gas as the most frequest word")
gas_curve_plot.axvline(dt.datetime(2019, 9, 21), color='green')
gas_curve_plot.axvline(dt.datetime(2018, 12, 17), color='green')
plt.show()
```
