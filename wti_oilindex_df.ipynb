```python
import pandas as pd
from pathlib import Path
```

```python
filepath = Path("Resources/wti_oilindex_texas.csv")
wti_index_df = pd.read_csv(filepath)
```

```python
wti_index_df
```

```python
wti_index_df.set_index('date', inplace=True)
wti_index_monthly = wti_index_df.index = pd.to_datetime(wti_index_df.index)
wti_index_monthly = wti_index_df.resample('1M').mean()
```

```python
wti_index_monthly
```

```python
wti_index_monthly.plot(kind='line', figsize=(20,10))
```

```python
wti_index_monthly.pct_change().plot(kind='line', figsize=(20,10))
```

```python
wti_index_monthly
```

```python
wti_index_monthly.reset_index(inplace=True)
wti_index_monthly['date'] = wti_index_monthly['date'].apply(lambda x: x.strftime('%Y-%m'))
wti_index_monthly.set_index('date', inplace=True)
```
